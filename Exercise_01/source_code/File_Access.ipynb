```python
image_name = 'images/kodim15_gray.png'

image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)

if image is None:
    print('Image ', image_name, ' not found')

image
plt.imshow(image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show()
cv2.imwrite('images/result_a.png', image) 
```

```python
image_name = 'images/kodim15.png'

image = cv2.imread(image_name, cv2.IMREAD_COLOR)

if image is None:
    print('Image ', image_name, ' not found')

image

color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(color_image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show() 
```

```python
image_name = 'images/kodim15_gray.png'

image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)

if image is None:
    print('Image ', image_name, ' not found')

image
plt.imshow(image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show()

print(image)
print(image.shape)
```

```python
#2.b The dimensions of the loaded data are defined by its size in pixels. We can see, that the immage is loaded as an array with the size of 512x768. Each value in the array, represents the
#greyscalevalue of one correlating pixel. We have 512 rows and 768 columns. We can either address a single pixelwith image[0,0] (first pixel in the upper left corner e.g.), but we can also select ranges.
#image[:100,0] gives us the first 100 of the first column, image[100:,0] gives us column 0 without the first 100 pixels, image[:,0] gives us access to the whole column. Ranges can used on both 
#dimensions toselect areas of the picture (e.g. an image[:50,:50] would give you a rectangle of the first 50x50 pixels in the upper left corner).
```

```python
image_name = 'images/kodim15.png'

image = cv2.imread(image_name, cv2.IMREAD_COLOR)

if image is None:
    print('Image ', image_name, ' not found')

image

color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(color_image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show() 

print(image)
print(color_image)
print(image.shape)
```

```python
#2.d The color version has an additional dimmension instead of a single value, to store its RGB values. The values are initially stored in the format (blue, green, red)/(b,g,r). 
#We can see the conversion from BGR to RGB happen in the arrays. The can be addressed the same way as the other dimensions (e.g. image[:50,:50,1] to get all green values of the square or e.g. image[:,:]
#to get all arrays of the 3rd dimension).
```
