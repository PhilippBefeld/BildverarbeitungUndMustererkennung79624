# Bibliotheken und Bilddarstellung

```python
import cv2
from matplotlib import pyplot as plt
```

## Einfache Grauwertbilder

Das Laden eines Bildes als Grauwertbild erfolgt durch 

```python
image_name = 'images/kodim15_gray.png'

image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)

if image is None:
    print('Image ', image_name, ' not found')

image

plt.imshow(image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show()
```

```python
image_name = 'images/kodim15_gray.png'

image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)

if image is None:
    print('Image ', image_name, ' not found')

image
image[:,:3] = (0)
image[:3,:] = (0)
image[:,765:] = (0)
image[508:,:] = (0)

plt.imshow(image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show()

cv2.imwrite('images/result_b.png', image) 
```

```python
image_name = 'images/kodim15_gray.png'

image = cv2.imread(image_name, cv2.IMREAD_GRAYSCALE)

if image is None:
    print('Image ', image_name, ' not found')

image

color_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGBA)

color_image[:,:10,3] = (127)
color_image[:10,:,3] = (127)
color_image[:,758:,3] = (127)
color_image[502:,:,3] = (127)

plt.imshow(color_image, cmap='gray')
_ = plt.xticks([]), plt.yticks([])
plt.show()

cv2.imwrite('images/result_c.png', color_image) 
```
